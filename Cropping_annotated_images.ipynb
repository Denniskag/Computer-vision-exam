```python
import os
import cv2
```

```python

input_folder = '/content/drive/MyDrive/Computer vision datasets/folder_B'
output_folder = os.path.join(input_folder, 'new_annotations4')
obj_names_path = '/content/drive/MyDrive/Computer vision datasets/names/obj.names'

if not os.path.exists(output_folder):
    os.makedirs(output_folder)
```

```python

def read_class_labels(obj_names_path):
    with open(obj_names_path, 'r') as file:
        class_labels = file.read().splitlines()

    class_labels[0], class_labels[1] = class_labels[1], class_labels[0]
    return class_labels


def process_image(image_path, annotation_path, class_labels):
    image = cv2.imread(image_path)
    height, width, _ = image.shape

    with open(annotation_path, 'r') as file:
        lines = file.readlines()
    for line in lines:
        class_id, x_center, y_center, box_width, box_height = map(float, line.strip().split())
        class_id = int(class_id)
        x_center *= width
        y_center *= height
        box_width *= width
        box_height *= height

        x_min = int(x_center - (box_width / 2))
        y_min = int(y_center - (box_height / 2))
        x_max = int(x_center + (box_width / 2))
        y_max = int(y_center + (box_width / 2))


        x_min = max(0, x_min)
        y_min = max(0, y_min)
        x_max = min(width, x_max)
        y_max = min(height, y_max)


        cropped_image = image[y_min:y_max, x_min:x_max]


        label = class_labels[class_id]
        label_folder = os.path.join(output_folder, label)
        if not os.path.exists(label_folder):
            os.makedirs(label_folder)

        base_filename = os.path.splitext(os.path.basename(image_path))[0]
        cropped_image_path = os.path.join(label_folder, f'{base_filename}_{label}_{x_min}_{y_min}.PNG')
        cv2.imwrite(cropped_image_path, cropped_image)
        print(f'Saved cropped image: {cropped_image_path}')


class_labels = read_class_labels(obj_names_path)


print("Class labels after swapping:", class_labels)


for filename in os.listdir(input_folder):
    if filename.endswith('.PNG'):
        image_path = os.path.join(input_folder, filename)
        annotation_path = os.path.join(input_folder, filename.replace('.PNG', '.txt'))

        if os.path.exists(annotation_path):
            process_image(image_path, annotation_path, class_labels)

print('Processing complete.')
```
